```python
import cv2
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt

# Загрузка изображения
uploaded = files.upload()
image_path = next(iter(uploaded))
image = cv2.imread(image_path)

# Преобразование в RGB (OpenCV загружает в BGR)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Преобразование в серый цвет
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Применение детектора контуров Кэнни
edges = cv2.Canny(gray_image, 100, 200)

# Найти контуры
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Создаем маску для рисования контуров
mask = np.zeros_like(image_rgb)

# Рисуем контуры на маске красным цветом
for contour in contours:
    cv2.drawContours(mask, [contour], -1, (255, 0, 0), 2)  # (255, 0, 0) - красный

# Накладываем маску на оригинальное изображение
contoured_image = cv2.addWeighted(image_rgb, 1, mask, 1, 0)

# Отображение изображения с контурами
plt.figure(figsize=(10, 10))
plt.imshow(contoured_image)
plt.title('Контуры объекта красным')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt

# Загрузка изображения
uploaded = files.upload()
image_path = next(iter(uploaded))
image = cv2.imread(image_path)

# Преобразуйте изображение в оттенки серого
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Примените детектор границ (например, Canny)
edges = cv2.Canny(gray, 100, 200)

# Найдите контуры
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Нарисуйте контуры на оригинальном изображении
# Сначала преобразуем изображение в RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
edges_rgb = cv2.cvtColor(edges, cv2.COLOR_BGR2RGB)

# Нарисуйте контуры на изображении в формате RGB
# (255, 0, 0) - это красный цвет в формате RGB
for contour in contours:
    cv2.drawContours(image_rgb, [contour], -1, (255, 0, 0), 2)  # 2 - толщина линии

# Сохраните или покажите изображение
plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.title('Контуры объекта красным')
plt.axis('off')
plt.show()

plt.figure(figsize=(10, 10))
plt.imshow(edges_rgb)
plt.title('Маска')
plt.axis('off')
plt.show()
```
